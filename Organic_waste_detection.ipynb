```python
from ultralytics import YOLO
import glob

# Load your trained model
model = YOLO("/content/drive/MyDrive/yolov9_runs/organic_waste_y95/weights/best.pt")

# List all val images
val_images = glob.glob("/content/drive/MyDrive/Organic Waste/images/val/*.jpg")

# Predict on val images and display them in Colab
for img_path in val_images:
    results = model.predict(
        source=img_path,  # image path
        show=True,        # display image with boxes in Colab
        save=False        # set True if you want to save the predicted images
    )
```
